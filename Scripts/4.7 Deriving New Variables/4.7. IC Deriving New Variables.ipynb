# Defining a user-defined function

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High-range product'
  else: return 'Not enough data'
```

## Using function on the dataframe

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

# If-Statements with For-Loops
## Step 1 - Creating busiest_day column

```python
# Checking the frequency
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creating a for-loop

result = []

for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
# Creating a new column for results
df_ords_prods_merged['busiest_day'] = result
```

```python
# Checking the frequency
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# Task Steps

## Step 2 - Creating a new column for busiest days in response to new requirements

```python
# Creating a for-loop

result_2 = []

for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0 or value == 1:
        result_2.append("Busiest days")
    elif value == 3 or value == 4:
        result_2.append("Least busy days")
    else:
        result_2.append("Regularly busy")
```

```python
# Adding results to the dataframe in a new columns
df_ords_prods_merged['busiest_days'] = result_2
```

```python
# Checking the output
df_ords_prods_merged.head()
```

## Step 3 - Check the values 

```python
# Checking the frequency
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

# Step 4 - Creating a new column for the busiest period of the day

```python
# Checking the frequency of values in the column
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### Since the client wants three different groups, 24 hours can be divided into 8 hour groups.

```python
# Creating a for-loop

result_3 = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_3.append("Most orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result_3.append("Average orders")
    else:
        result_3.append("Fewest orders")
```

```python
# Adding results to the dataframe in a new column
df_ords_prods_merged['busiest_period_of_day'] = result_3
```
